### Popularity based Recommender 

- Rely on history of purchase, previously liked, clicked etc 
- Based on whats most popular and are thus not personalized for particular user 
- Often used in systems where they want to recommend what is popular generally such as new sites etc. 


```python
import numpy as np 
import pandas as pd
```

```python
ratings = pd.read_csv('rating_final.csv')
cusine = pd.read_csv('chefmozcuisine.csv')
```

```python
ratings.head()
```

```python
cusine.head()
```

```python
ratings_count = pd.DataFrame(ratings.groupby('placeID').count())
ratings_count.sort_values('rating',ascending=False).head(5)
```

```python
top_5 = ratings_count.sort_values('rating',ascending=False).head(5).index.tolist()
most_rated = pd.DataFrame(top_5,index=np.arange(5),columns=['placeID'])
summary = pd.merge(most_rated,cusine,on='placeID')
summary
```
